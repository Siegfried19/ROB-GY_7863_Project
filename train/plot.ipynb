```python

import pandas as pd
import matplotlib.pyplot as plt

import numpy as np


LOG_PATH = "go2_moon_log.csv"  # 修改为你的路径
df = pd.read_csv(LOG_PATH)
print(f"已读取文件：{LOG_PATH}，共 {len(df)} 条数据")

# 显示前几行看看格式
df.head()


LOG_PATH = "go2_moon_reference_log.csv"  # 修改为你的路径
df_ref = pd.read_csv(LOG_PATH)
print(f"已读取文件：{LOG_PATH}，共 {len(df)} 条数据")

# 显示前几行看看格式

start_index = 0
end_index = 2300
x = np.arange(len(df))[start_index:end_index]/200
df_ref.head()

# window_size = 10  # 可调：越大越平滑
# df = df.rolling(window=window_size, center=True).mean()
```

```python
plt.figure(figsize=(14, 10))


print(len(df["time"][start_index:]))
plt.title("Reference Joint Angles (rad)")
for i in range(12):
    if f"qpos_{i}" in df.columns:
        plt.plot(x , df_ref[f"ref_angle_{i}"][start_index:end_index], label=f"ref_q{i}")
plt.xlabel("Time (s)")
plt.ylabel("Angle (rad)")
plt.legend(ncol=6, fontsize=7)
plt.grid(True)
```

```python

plt.title("Joint Comparation")

plt.plot(x, df[f"qpos_{7}"][start_index:end_index], label=f"qpos_{0}")
plt.plot(x, df_ref[f"ref_angle_{7}"][start_index:end_index], label=f"ref_qpos{0}")

plt.xlabel("Time (s)")
plt.ylabel("Joint")
plt.legend(ncol=6, fontsize=7)
plt.grid(True)
```

```python

plt.figure(figsize=(14, 10))


plt.title("Joint Angles (rad)")
for i in range(12):
    if f"qpos_{i}" in df.columns:
        plt.plot(x, df[f"qpos_{i}"][start_index:end_index], label=f"q{i}")
plt.xlabel("Time (s)")
plt.ylabel("Angle (rad)")
plt.legend(ncol=6, fontsize=7)
plt.grid(True)
```

```python


plt.title("Orientation (Euler angles)")
plt.plot(x, df["roll"][start_index:end_index], label="roll")
plt.plot(x, -df["pitch"][start_index:end_index], label="pitch")

plt.xlabel("Time (s)")
plt.ylabel("Angle (rad)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python

from scipy.ndimage import gaussian_filter1d
plt.figure(figsize=(14, 10))
print(len(df["time"][start_index:]))
plt.title("Predicted Torque")
for i in range(12):
    if f"action_{i}" in df.columns:
        smooth_tau = gaussian_filter1d(df[f"action_{i}"][start_index:end_index], sigma=3)
    
        plt.plot(x,smooth_tau , label=f"predited_q{i}")
plt.xlabel("Time (s)")
plt.ylabel("Action")
plt.legend(ncol=6, fontsize=7)
plt.grid(True)
```
